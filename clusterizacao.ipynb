```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd
```

```python
df = pd.read_excel('base.xlsx')
```

```python
df
```

```python
# Selecionar as colunas numéricas para a clusterização
cols_to_cluster = ['Nota Pergunta 1', 'Nota Pergunta 2', 'Sacl Ges Resp Ouvidoria']

# Filtrar o DataFrame para as colunas selecionadas
df_cluster = df[cols_to_cluster].copy()
df_cluster
```

```python
# Padronizar os dados numéricos
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_cluster)
df_scaled
```

```python
# Criar e ajustar o modelo KMeans com 3 clusters (como exemplo inicial)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df_scaled)
kmeans
```

```python
# Adicionar os rótulos dos clusters ao dataframe original
df['Cluster'] = kmeans.labels_

# Calcular a métrica de silhueta para avaliar a qualidade dos clusters
silhouette_avg = silhouette_score(df_scaled, kmeans.labels_)

silhouette_avg
```
